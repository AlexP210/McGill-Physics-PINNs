```python
from tensorflow.keras import Sequential, Model, layers, Input
from tensorflow import keras
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
# Random x data, 10 scalar inputs
x = np.linspace(0, 10, 10)[:,np.newaxis]
# Random y-values, 10 scalar outputs
y_actual = x**2 + np.random.random(10)[:, np.newaxis]
# Check the shapes
print(x.shape)
print(y_actual.shape)
# Plot the data
plt.plot(x, y_actual)
plt.show()
```

```python
# Create a tensorflow model
model = Sequential([keras.Input(1), layers.Dense(10), layers.Dense(1, activation="ReLU")])
y_predicted = model(x)
plt.plot(x, y_actual)
plt.plot(x, y_predicted)
plt.show()
```

```python
# Train the model, and show the fit
model.summary()
model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.1), loss='mean_absolute_error')
model.fit(x, y_actual, epochs=100, verbose=False)
y_predicted = model.predict(x)
plt.plot(x, y_actual)
plt.plot(x, y_predicted)
plt.show()
```
